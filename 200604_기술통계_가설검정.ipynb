### 기술통계
**: 연속형 데이터를 수치로 요약하여 데이터의 전반적인 특성을 파악**

모집단 : 조사될 수 있는 모든 데이터들을 말한다.

표본집단 : 모집단에서 일부 데이터들을 선정한 집단을 말한다.

평균 : 모든 값의 합을 데이터의 개수로 나눈 값<br>
(계산이 쉽지만 이상치에 민감하여 제거해줘야 함)

중앙값 : 데이터를 크기순으로 정렬했을 때 중앙에 있는 값<br>
(평균과 달리 이상치에 영향을 받지 않음)

분산 : 각 데이터에서 평균을 뺀 뒤 제곱한 값을 모두 더한 후 데이터 개수로 나눈 값

표준편차 : 데이터가 얼마나 넓게 퍼져 있는가를 나타내는 값(분산에 루트를 씌운 값)

표준오차 : 표본집단들 평균에 대한 표준편차

최빈값 : 데이터 중 가장 여러번 나타난 값

최대값 : 데이터에서 가장 큰 값

최소값 : 데이터에서 가장 작은 값

사분위수 : 데이터를 4개의 부분으로 나눈 값
- 제 1사분위수(25%) : 데이터의 25%가 이 값보다 작거나 같음
- 제 2사분위수(50%) : 데이터의 50%가 이 값보다 작거나 값음
- 제 3사분위수(75%) : 데이터의 75%가 이 값보다 작거나 같음

### 가설검정

귀무가설 : 일반적으로 맞다고 가정하는 가설

대립가설 : 새로운 주장 또는 맞다고 증명하려는 가설

신뢰수준 : 가설을 검정할 때 얼마나 검정할 것인지를 결정하는 수준<br>
일반적으로 95% 정도의 신뢰수준을 사용

유의수준 : 가설을 검정할 때 신뢰수준을 벗어나는 수준<br>
1 - 신뢰수준

p-value(유의확률) : 가설을 결정할 수 있는 확률

검정통계량 : 가설을 검정할 목적으로 사용되는 통계량

검정력 : 대립가설이 사실일 때, 이를 사실로서 결정할 확률

자유도 : 모집단에 대한 정보를 주는 독립적인 자료의 수
