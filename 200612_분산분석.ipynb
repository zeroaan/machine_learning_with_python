### t-검정
: 두 집단의 평균이 유의한 차이가 있는지 검증하는 방법

독립성, 정규성, 등분산성을 만족할 경우 사용 가능하다.

<br>

**독립성이란 : 한 집단의 측정값이 다른 집단에 의해 영향을 받지 않는 것을 의미한다.**<br>

ex) A실험, B실험이 있을 때, A실험을 하는 집단과 B실험을 하는 집단이 따로 측정하는 것이 독립되었다고 한다.<br>
만약 한 집단이 A실험을 하고 난 후 B실험을 이어서 한다면 이 경우 독립되지 않았다고 한다.

<br>

**정규성이란 : 측정값이 한쪽으로 치우쳐져 있지 않고 평균을 중심으로 이루어지는 것을 말하며 일반적으로 측정값이 30개 이상이면 정규성을 띤다고 가정할 수 있으며, 30개 이하일 경우에는 Shapiro test로 정규성을 증명해야 한다.**<br>

정규성 검정의 가설은 다음과 같이 설정한다.<br>

귀무가설 : 정규성을 만족한다.(해당 분포가 정규분포와 차이가 없다.)<br>
대립가설 : 정규성을 만족하지 않는다.(해당 분포가 정규분포와 차이가 있다.)<br>

결과 p-value가 0.05보다 크면 귀무가설이 채택되어 정규성을 갖는 것으로 판단한다.

<br>

**등분산성이란 : 두 집단의 분산이 같음을 의미한다. levene 검정으로 증명해야 한다.**<br>

등분산성의 가설은 다음과 같이 설정한다.<br>

귀무가설 : 등분산성을 만족한다.(해당 분산들이 차이가 없다.)<br>
대립가설 : 등분산성을 만족하지 않는다.(해당 분산들이 차이가 있다.)<br>

결과 p-value가 0.05보다 크면 귀무가설이 채택되어 등분산성을 만족한다고 판단한다.<br>
하지만 등분산성이 만족되지 않더라도 t-검정에서 equal_var=False 옵션을 추가하여 진행할 수 있다.
<br>

<br><br><br>

단일표본 t-검정<br>
: 하나의 집단에서 구한 평균이 어떤 특정 값과 유의한 차이가 있는지 검증하는 방법<br>

귀무가설 : 한 집단의 평균이 특정 값과 차이가 없다.<br>
대립가설 : 한 집단의 평균이 특정 값과 차이가 있다.

ex) 컴퓨터 공학과 남학생들의 키를 조사하였다. 대한민국 남자 키 평균으로 알려져 있는 174cm와 차이가 있는지 검증하자.

<br>

독립표본 t-검정<br>
: 서로 다른 두 집단의 평균이 유의한 차이가 있는지 검증하는 방법<br>

귀무가설 : 두 집단의 평균이 차이가 없다.<br>
대립가설 : 두 집단의 평균이 차이가 있다.

ex) 성별에 따른 성적의 차이가 있는지 검증하자.

<br>

대응표본 t-검정<br>
: 한 집단이 서로 다른 두 조건에서 진행한 표본의 평균이 유의한 차이가 있는지 검증하는 방법<br>
(서로 다른 두 조건이란 시간의 전, 후 차이를 말함)<br>

귀무가설 : 전과 후의 평균의 차이가 없다.<br>
대립가설 : 전과 후의 평균의 차이가 있다.

ex) 세미나 이전의 성적과 이후의 성적이 차이가 있는지 검증하자.

<br>

<br><br><br>

### 분산분석
: 두 개 이상의 집단을 비교할 때 사용하는 검정 방법

t-test를 여러 번 하면 되는데 분산분석을 하는 이유?<br>
: t-test에서 일반적으로 95%의 신뢰도를 갖고 진행을 하는데, 만약 두 번 진행했다고 하면 95% * 95%가 신뢰도가 되므로 신뢰도가 감소한다.
<br>

마찬가지로 분산분석도 독립성, 정규성, 등분산성을 만족해야한다.<br>


<br><br>

일원 분산분석<br>
: 독립변수 1개, 종속변수 1개 일 때 집단 간의 평균이 유의한 차이가 있는지 검정<br>
ex) 학과에 따른 성적 비교
<br>

다원 분산분석<br>
: 독립변수 n개, 종속변수 1개 일 때 집단 간의 평균이 유의한 차이가 있는지 검정<br>
ex) 학과와 공부방법에 따른 성적 비교
<br>

다변량 분산분석<br>
: 독립변수 1개이상, 종속변수 2개이상 일 때 집단 간의 평균이 유의한 차이가 있는지 검정<br>
ex) 학과에 따른 성적과 공부시간 비교
<br>
<br>

- 귀무가설 : 집단들의 평균이 차이가 없다.
- 대립가설 : 집단들의 평균에 적어도 하나 차이가 있다.

<br><br><br>

### 사후 검증
: 분산분석은 전체 집단에서 차이가 있는지를 검정하는 방법이지만, 어떤 집단에서 차이가 있는지는 알지 못한다. 그렇기 때문에 분산분석 결과 유의미하다는 결론이 나왔을 때, 어떤 집단에서 평균 차이가 나는지 알기 위해 사후 검증을 한다.
<br>

대표적인 사후검증 방법으로 Duncan, Tukey, Scheffe, Bonferroni가 있다.
<br>

1. Duncan : 작은 차이에도 차이가 있다라고 한다.<br>
분석 대상의 집단을 단순히 나누고 싶을때 사용
<br>

2. Scheffe : 확실한 차이가 있어야 차이가 있다라고 한다.<br>
통계적으로 유의한 차이를 보기 힘듬
<br>

3. Tukey : 집단의 수가 동일할 경우 사용한다.<br>
Duncan과 Scheffe의 중간 정도의 성격을 가지고 있음
<br>

4. Bonferroni : 집단의 수가 동일하지 않을 경우 사용한다.<br>
Duncan과 Scheffe의 중간 정도의 성격을 가지고 있음
<br>


집단을 분리하고자 하는 기준에 따라 선택하면 된다.

<br><br><br>

### 일원분산분석 실습

```python
import warnings
warnings.filterwarnings('ignore')
```

```python
import pandas as pd

df = pd.read_excel('value_data.xlsx')
df
```

```python
df = df[['value', 'group']]
df.head()
```

```python
X_test = df[df.group == 'X'].value
Y_test = df[df.group == 'Y'].value
Z_test = df[df.group == 'Z'].value
```

```python
from scipy.stats import shapiro

print("X:", shapiro(X_test)[1], 
      "\nY:", shapiro(Y_test)[1], 
      "\nZ:", shapiro(Z_test)[1])
```

```python
from scipy.stats import levene

print("p-value:", levene(X_test, Y_test, Z_test)[1])
```

<br><br><br>

- 귀무가설 : 집단들의 평균이 차이가 없다.
- 대립가설 : 집단들의 평균에 적어도 하나 차이가 있다.

```python
from statsmodels.formula.api import ols
from statsmodels.stats.anova import anova_lm

model = ols('value ~ C(group)', df).fit()
anova_lm(model)
```

```python
df.groupby(['group']).agg(len)
```

```python
from statsmodels.stats.multicomp import pairwise_tukeyhsd

hsd = pairwise_tukeyhsd(df['value'], df['group'], alpha=0.05)
hsd.summary()
```

### 이원분산분석 실습

```python
import pandas as pd

df = pd.read_excel('value_data.xlsx')
df = df[['value', 'group', 'method']]
df.head()
```

```python
a_test = df[df.method == 'a'].value
b_test = df[df.method == 'b'].value
c_test = df[df.method == 'c'].value
d_test = df[df.method == 'd'].value
e_test = df[df.method == 'e'].value
```

```python
from scipy.stats import shapiro

print("a:", shapiro(a_test)[1], 
      "\nb:", shapiro(b_test)[1], 
      "\nc:", shapiro(c_test)[1], 
      "\nd:", shapiro(d_test)[1], 
      "\ne:", shapiro(e_test)[1])
```

```python
from scipy.stats import levene

print("p-value:", levene(a_test, b_test, c_test, d_test, e_test)[1])
```

<br><br><br>

- 귀무가설 : 집단들의 평균이 차이가 없다.
- 대립가설 : 집단들의 평균에 적어도 하나 차이가 있다.

```python
df.groupby(['method', 'group']).agg(len)
```

```python
from statsmodels.formula.api import ols
from statsmodels.stats.anova import anova_lm

model = ols('value ~ C(method) * C(group)', df).fit()
anova_lm(model, typ=3)
```

```python
df.groupby(['group']).agg(len)
```

```python
from statsmodels.stats.multicomp import pairwise_tukeyhsd

hsd = pairwise_tukeyhsd(df['value'], df['group'], alpha=0.05)
hsd.summary()
```

```python
df.groupby(['method']).agg(len)
```

```python
from statsmodels.sandbox.stats.multicomp import MultiComparison
import scipy.stats

comp = MultiComparison(df['value'], df['method'])

result = comp.allpairtest(scipy.stats.ttest_ind, method='bonf')
result[0]
```

### 다변량 분산분석

```python
import pandas as pd

df = pd.read_excel('value_data.xlsx')
df = df[['value', 'group', 'day']]
df.head()
```

```python
from statsmodels.multivariate.manova import MANOVA

maov = MANOVA.from_formula('value + day ~ C(group)', df)
print(maov.mv_test())
```

```python
from statsmodels.stats.multicomp import pairwise_tukeyhsd

hsd = pairwise_tukeyhsd(df['day'], df['group'], alpha=0.05)
hsd.summary()
```

### 대장암 데이터셋 이용

#### 일원분산분석

```python
import pandas as pd

df2 = pd.read_excel('NIPA_EMR.xlsx')
df2.head()
```

```python
df2['tumor_loc'] = df2['Tumor_location_pathology'].astype(str)

tumor_loc = df2.tumor_loc
tumor_loc
```

```python
op_time = df2.op_time
op_time
```

```python
data2 = {'tumor_loc': tumor_loc, 'op_time': op_time}

op_df = pd.DataFrame(data2)
op_df
```

```python
from statsmodels.formula.api import ols
from statsmodels.stats.anova import anova_lm

model = ols('op_time ~ C(tumor_loc)', op_df).fit()
anova_lm(model)
```

```python
op_df.groupby(['tumor_loc']).agg(len)
```

```python
from statsmodels.sandbox.stats.multicomp import MultiComparison
import scipy.stats

comp = MultiComparison(op_df['op_time'], op_df['tumor_loc'])

result = comp.allpairtest(scipy.stats.ttest_ind, method='bonf')
result[0]
```
